```python
# ✅ 1. 필요한 라이브러리 설치 (Colab에서는 기본 제공됨)
import pandas as pd
import openai
from google.colab import files

# ✅ 2. CSV 파일 업로드
uploaded = files.upload()

# ✅ 3. CSV 파일 읽기 (E01_04.csv를 기준으로)
df = pd.read_csv("E01_04.csv")

# ✅ 4. 구절 형식으로 구성
verses = []
for _, row in df.iterrows():
    book = str(row['book']).strip()
    chapter = int(row['chapter'])
    verse = int(row['verse'])
    text = str(row['text']).strip()

    verses.append(f"{book} {chapter}:{verse} - {text}")

# ✅ 5. 프롬프트 구성
prompt = (
    "아래는 신약 성경에서 'Joseph'이라는 이름이 나오는 구절들입니다.\n"
    "각 구절을 분석하여 Joseph이라는 이름이 서로 다른 몇 명의 인물로 등장하는지 구분해 주세요.\n"
    "가능하다면 구절들을 분류해서 어떤 구절이 어떤 인물로 보이는지도 알려주세요.\n\n"
    "구절 목록:\n" + "\n".join(verses)
)

# ✅ 6. OpenAI API 키 설정
openai.api_key = "API KEY"  # ← 여기에 본인의 OpenAI 키를 입력하세요

# ✅ 7. GPT 호출
response = openai.chat.completions.create(
    model="gpt-4",  # 또는 "gpt-3.5-turbo"
    messages=[
        {"role": "system", "content": "당신은 성경 인물 분석 전문가입니다."},
        {"role": "user", "content": prompt}
    ],
    temperature=0.1
)

# ✅ 8. 결과 출력
print(response.choices[0].message.content)
```
